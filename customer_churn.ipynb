```python
import pandas as pd # Data Cleaning
import numpy as np # Array Management
import matplotlib.pyplot as plt # Visualizations
```

```python
customer_data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/customer_churn.csv')
```

```python
# Step 1: Impute missing numerical values with mean
numeric_columns_with_missing = [
    "Avg Monthly Long Distance Charges",
    "Avg Monthly GB Download"
]
for col in numeric_columns_with_missing:
    customer_data[col].fillna(customer_data[col].mean(), inplace=True)
```

```python
# Step 2: Impute missing categorical values with mode
categorical_columns_with_missing = [
    "Multiple Lines",
    "Internet Type",
    "Online Security",
    "Online Backup",
    "Device Protection Plan",
    "Premium Tech Support",
    "Streaming TV",
    "Streaming Movies",
    "Streaming Music",
    "Unlimited Data"
]
for col in categorical_columns_with_missing:
    customer_data[col].fillna(customer_data[col].mode()[0], inplace=True)
























```

```python
# Step 3: Drop columns with excessive missing values
columns_to_drop = ["Churn Reason"]
customer_data.drop(columns=columns_to_drop, inplace=True)
```

```python
# Save the final dataset
customer_data.to_csv('processed_customer_churn.csv', index=False)
```
